```python
input_drug = 'avelox'
drugname = new_df.loc[new_df['Drug'] == input_drug]
```

```python
pd.set_option('display.max_colwidth', -1)

print(drugname.Side_Effects)
```

```python
#Count frequency for 'All comments'
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Build frequency distance matrix
txt = drugname.Side_Effects.str.replace(r'\|', ' ').str.cat(sep=' ')
tokenized_words = nltk.tokenize.word_tokenize(txt)
stop_words = stopwords.words('english')
custom_stopwords = ['may','occur','far','drug','product','change','extremely','extreme','dose','much','really','felt','early','give','began','stop','stopped','feel','feeling','go','going','even','pill','capsule','able','along','easily','seemed','started','made','got','need','help','helped','one','two','three','put','experience','experienced','am','pm','hour','hours','hr','hrs','less','more','lot','use','around','gets','taken','medication','medicine','taking','adjusts','body','tell','still','side','effects','became','become','terrible','worse','caused','cause','problems','problem','none','n\'t','mg','mgs','amount','involved', 'Name', 'dtype', 'object', 'series', 'Word','occured','noticed','doctor','pharmacist','promptly','incident','effect','effected','affected','time','take','first','last','\'ve','every','initially','gradually','using','next','day','days','week','weeks','month','months','very','treatment','affect','within']
stop_words.extend(custom_stopwords)
stop_words.append(input_drug)
words = [word for word in tokenized_words if word not in stop_words]
word_dist = nltk.FreqDist(words)

# # Words with higher frequency
#word_dist.plot(30,cumulative=False)

# The top 100 words
rslt = pd.DataFrame(word_dist.most_common(30), columns=['Word', 'Frequency'])


text = " ".join(review for review in rslt.Word)
#print(text)
#display(rslt)
if len(rslt) == 0:
    text = 'None'

wordcloud1 = WordCloud(max_font_size=50, max_words=12, background_color="white").generate(text)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud1, interpolation="bilinear")
plt.axis("off")
plt.show()
```
